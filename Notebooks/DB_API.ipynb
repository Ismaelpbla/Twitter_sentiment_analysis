```python
# Importamos las librerias necesarias

import pandas as pd
import sqlite3
```

```python
# Hacemos una conexi√≥n a una base de datos que llamamos my_database

connt = sqlite3.connect('Data/API_database.db')

ct = connt.cursor() #Instanciamos un cursor
```

```python
# Creamos una tabla llamada tweets
ct.execute(
    """CREATE TABLE IF NOT EXISTS tweets(
        author_id VARCHAR(255),
        text TEXT NOT NULL,
        created_at DATE,
        id_tweet VARCHAR(255) PRIMARY KEY,
        retweet_count INTEGER,
        reply_count INTEGER,
        like_count INTEGER,
        quote_count INTEGER
        )"""
    )

# Creamos una tabla llamada users
ct.execute(
    """CREATE TABLE IF NOT EXISTS users(
    author_id VARCHAR(255) PRIMARY KEY,
    name VARCHAR(255),
    username VARCHAR(255)
    )"""
    )


sql = '''SELECT * FROM tweets'''
ct.execute(sql)
```

```python
df = pd.read_csv('Data\clean1_data.csv', index_col=0)

df = df.rename({'id':'id_tweet'},axis=1)

df.head(50)
```

```python
df_authors = pd.read_csv('Data/author.csv', index_col=0)


df_authors.rename({'id':'author_id'}, axis=1)
```

```python

# Insert DataFrame recrds one by one.
for i,row in df.iterrows():
    task = (row[0], row[1], row[2], row[3], row[4], row[5], row[6])
    sql = ''' INSERT INTO tweets(author_id, text, created_at, id_tweet, retweet_count, like_count,quote_count)
              VALUES(?,?,?,?,?,?,?) '''
    cur = connt.cursor()
    cur.execute(sql, task)
    connt.commit()
```

```python

# Insert DataFrame recrds one by one.
for i,row in df_authors.iterrows():
    task = (row[0], row[1], row[2])
    sql = ''' INSERT INTO users(author_id,name,username)
              VALUES(?,?,?) '''
    cur = connt.cursor()
    cur.execute(sql, task)
    connt.commit()
```

```python
connt.close()
```
