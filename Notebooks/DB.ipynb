```python
import pandas as pd

import sqlite3
```

```python

connt = sqlite3.connect('Data/my_database.db')

ct = connt.cursor()
```

```python
ct.execute(
    """CREATE TABLE IF NOT EXISTS tweets(
        id VARCHAR(255) PRIMARY KEY,
        tweet TEXT NOT NULL,
        date DATE, 
        user_id VARCHAR(255),
        retweets_count INTEGER,
        replies_count INTEGER,
        likes_count INTEGER
        )"""
    )

ct.execute(
    """CREATE TABLE IF NOT EXISTS users(
    user_id VARCHAR(255) PRIMARY KEY,
    name VARCHAR(255),
    username VARCHAR(255)
    )"""
    )

sql = '''SELECT * FROM tweets'''
ct.execute(sql)
```

```python
df = pd.read_csv('Data\clean_data.csv', index_col=0)

df

```

```python
df_tweets = df.drop(['username','name'], axis=1)

# Insert DataFrame recrds one by one.
for i,row in df_tweets.iterrows():
    task = (row[0], row[1], row[2], row[3], row[4], row[5], row[6])
    sql = ''' INSERT INTO tweets(id, date, user_id, tweet, replies_count, retweets_count, likes_count)
              VALUES(?,?,?,?,?,?,?) '''
    cur = connt.cursor()
    cur.execute(sql, task)
    connt.commit()
```

```python
df_user = df.loc[:, ['user_id','username','name']].drop_duplicates()

# Insert DataFrame recrds one by one.
for i,row in df_user.iterrows():
    task = (row[0], row[1], row[2])
    sql = ''' INSERT INTO users(user_id,username,name)
              VALUES(?,?,?) '''
    cur = connt.cursor()
    cur.execute(sql, task)
    connt.commit()
```

```python
connt.close()
```
